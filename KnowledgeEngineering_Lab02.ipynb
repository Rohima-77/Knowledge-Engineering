```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()
G.add_nodes_from([1,2,3,4,5,6,7,8])
G.add_edges_from([(1,2),(1,3),(2,4),(2,5),(3,6),(3,7),(4,8),(5,8),(6,8),(7,8)])
plt.figure(figsize=(10,10))
nx.draw_networkx(G, with_labels=True, node_size=2000,font_size=14)
```

```python
#adjacency matrix
# Relation between node & node
adj_matrix = nx.adjacency_matrix(G).todense()
print(adj_matrix)
```

```python
import pandas as pd
nodes = list(G.nodes())
adj_matrix_df = pd.DataFrame(adj_matrix, index=nodes, columns=nodes)
print("Adjacency Matrix with Labels:")
display(adj_matrix_df)
```

```python
# Incidance matrix
# Relation between node and edges
incidance = nx.incidence_matrix(G, oriented=True).todense()
print(incidance)
```

```python
from networkx.classes import edges
import pandas as pd
nodes = list(G.nodes())
edges = list(G.edges())
incidance_df = pd.DataFrame(incidance, index=nodes, columns=edges)
print("Incidance Matrix with Labels:")
display(incidance_df)
```

```python
from numpy import expm1
G = nx.Graph()
edges = [(1,2,"e1"),
         (1,3,"e2"),
         (2,4,"e3"),
         (2,5,"e4"),
         (3,6,"e5"),
         (3,7,"e6"),
         (4,8,"e7"),
         (5,8,"e8"),
         (6,8,"e9"),
         (7,8,"e10")]
for u,v,lab in edges:
    G.add_edge(u,v,label=lab)
pos = nx.spring_layout(G)
plt.figure(figsize=(10,10))
nx.draw_networkx(G, pos, with_labels=True, node_size=2000,font_size=14)
edge_labels = nx.get_edge_attributes(G,'label')
nx.draw_networkx_edge_labels(G,pos,edge_labels=edge_labels,font_size=14)
plt.title("Weighted Graph with Edge Labels")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
data = [
    ("A","Inception",5),
    ("A","Avatar",4),
    ("B","Inception",4),
    ("B","Titanic",5),
    ("C","Titani",4),
    ("C","Avatar",5),
    ("D","Inception",1),
    ("D","Joker",5),
    ("E","Joker",4),
    ("E","Titanic",1)


]
```

```python
df = pd.DataFrame(data, columns=["User","Movie","Rating"])
print(df)
```

```python
B=nx.Graph()
for person,movie,rating in data:
    B.add_node(person,bipartite=0)# Right side node = 0
    B.add_node(movie,bipartite=1)# Left side node = 1
    B.add_edge(person,movie,weight=rating)
nx.draw(B,with_labels=True,font_weight='bold')
plt.show()
```

```python
people_nodes=[n for n,d in B.nodes(data=True) if d['bipartite']==0]
movie_nodes=[n for n,d in B.nodes(data=True) if d['bipartite']==1]
#layout for Bipertite

pos = dict()
pos.update((n, (0, i)) for i, n in enumerate(people_nodes))
pos.update((n, (1, i)) for i, n in enumerate(movie_nodes))

#Draw nodes
```
