**Question 1. Create an Undirected Graph**

```python
import networkx as nx

G = nx.Graph()

G.add_edges_from([
    ("Ayesha", "Rahim"),
    ("Ayesha", "Karim"),
    ("Rahim", "Sadia"),
    ("Karim", "Nabil"),
    ("Sadia", "Nabil"),
    ("Nabil", "Tania"),
    ("Tania", "Ayesha")
])

print("Undirected Graph Created Successfully.")
```

**2.Draw the graph**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

pos = nx.spring_layout(G, seed=42)

nx.draw(
    G,
    pos,
    with_labels=True,
    node_color="skyblue",
    node_size=2500,
    font_size=10,
    font_weight="bold"
)

plt.title("Friendship Graph")
plt.show()
```

**Question 3. Print all Nodes, Edges, Number of Nodes and Edges**

```python
print("All Nodes:")
print(G.nodes())

print("All Edges:")
print(G.edges())

print("Total Number of Nodes:")
print(G.number_of_nodes())

print("Total Number of Edges:")
print(G.number_of_edges())
```

**Question 4. Which Student has the Highest Number of Connections?**

```python
degrees = dict(G.degree())

print("Degree of Each Student:")
print(degrees)

highest = max(degrees, key=degrees.get)

print("\nStudent with Highest Connections:")
print(highest)

print("Total Connections:", degrees[highest])
```

**Question 5. Remove Rahim–Karim Friendship and Redraw**

```python
if G.has_edge("Rahim", "Karim"):
    G.remove_edge("Rahim", "Karim")
else:
    print("Rahim and Karim are not friends.")
plt.figure(figsize=(8,6))

nx.draw(
    G,
    pos,
    with_labels=True,
    node_color="lightgreen",
    node_size=2500,
    font_weight="bold"
)

plt.title("Graph after Removing Friendship")
plt.show()
```

**Question 6. Add Farhan and New Friendships**

```python
G.add_node("Farhan")

G.add_edge("Farhan","Karim")
G.add_edge("Farhan","Tania")
plt.figure(figsize=(8,6))

pos = nx.spring_layout(G, seed=42)

nx.draw(
    G,
    pos,
    with_labels=True,
    node_color="orange",
    node_size=2500,
    font_weight="bold"
)

plt.title("Updated Graph")
plt.show()
```

**Question 7:Convert into Weigted Directed Graph**

```python
DG = nx.DiGraph()

DG.add_weighted_edges_from([
    ("Ayesha","Rahim",3),
    ("Ayesha","Karim",5),
    ("Rahim","Sadia",2),
    ("Karim","Nabil",4),
    ("Sadia","Nabil",6),
    ("Nabil","Tania",3),
    ("Tania","Ayesha",7)
])

plt.figure(figsize=(8,6))

pos = nx.spring_layout(DG, seed=42)

nx.draw(
    DG,
    pos,
    with_labels=True,
    node_color="pink",
    node_size=2500,
    arrows=True
)

labels = nx.get_edge_attributes(DG, "weight")

nx.draw_networkx_edge_labels(
    DG,
    pos,
    edge_labels=labels
)

plt.title("Weighted Directed Graph")
plt.show()
```

**Question 8:Prove Handshake Theorem**

```python
degrees = dict(G.degree())

print("Degrees:")
print(degrees)

sum_degree = sum(degrees.values())

edges = G.number_of_edges()

print("\nSum of Degrees =", sum_degree)
print("2 × Number of Edges =", 2 * edges)

if sum_degree == 2 * edges:
    print("\nHandshake Theorem Proved.")
else:
    print("\nHandshake Theorem Not Proved.")
```
