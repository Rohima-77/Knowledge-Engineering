**Question:1 Draw the Graph**

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create Graph
G = nx.Graph()

# Add Edges
G.add_edges_from([
    ("A","B"),
    ("A","C"),
    ("B","D"),
    ("C","D"),
    ("C","E"),
    ("D","E")
])

# Draw Graph
plt.figure(figsize=(6,5))

nx.draw(
    G,
    with_labels=True,
    node_color="skyblue",
    node_size=2000,
    font_size=12,
    font_weight="bold"
)

plt.title("Social Network Graph")
plt.show()
```

**Question 2. Adjacency Matrix**

```python
import pandas as pd

vertices = ["A","B","C","D","E"]

adj = nx.adjacency_matrix(G)

adj_df = pd.DataFrame(
    adj.toarray(),
    index=vertices,
    columns=vertices
)

print(adj_df)
```

**Question 3. Incidence Matrix**

```python
inc = nx.incidence_matrix(G, oriented=False)

inc_df = pd.DataFrame(
    inc.toarray().astype(int),
    index=vertices,
    columns=[f"e{i+1}" for i in range(G.number_of_edges())]
)

print(inc_df)
```

**Question 4(i). Degree of Each Node**

```python
degrees = dict(G.degree())

print("Degree of each node:")

for node, degree in degrees.items():
    print(node, ":", degree)
```

**Question 4(ii). Node with Highest Degree**

```python
highest = max(degrees, key=degrees.get)

print("Node with Highest Degree:", highest)
print("Degree:", degrees[highest])
```

**Question 5(i). Directed or Undirected?**

```python
if G.is_directed():
    print("Directed Graph")
else:
    print("Undirected Graph")
```

**Question 5(ii). Weighted or Unweighted?**

```python
weighted = nx.is_weighted(G)

if weighted:
    print("Weighted Graph")
else:
    print("Unweighted Graph")
```

**Question 6. Dictionary-based Adjacency List**

```python
graph = {
    "A": ["B", "C"],
    "B": ["A", "D"],
    "C": ["A", "D", "E"],
    "D": ["B", "C", "E"],
    "E": ["C", "D"]
}

print("Adjacency List:")

for node in graph:
    print(node, ":", graph[node])
```
